### Section 0: STA waveNumbers

```python
def get_primary(path_stacks_folder):
    pattern = os.path.join(path_stacks_folder, "*STA__1S*")
    for p in sorted(glob.glob(pattern)):
        sta_path = Path(p)
        sta_product = BiomassProduct.BiomassProductSTA(sta_path)
        if sta_product.isCoregistrationPrimary:
            print(f"[INFO] Found primary STA: {sta_path}")
            return sta_path
def get_first_secondary(path_stacks_folder, primary_path):
    pattern = os.path.join(path_stacks_folder, "*STA__1S*")

    for p in sorted(glob.glob(pattern)):
        sta_path = Path(p)

        # salta il primary
        if sta_path == primary_path:
            continue

        sta_product = BiomassProduct.BiomassProductSTA(sta_path)

        if not sta_product.isCoregistrationPrimary:
            print(f"[INFO] Found SECONDARY STA: {sta_path.name}")
            return sta_path

    raise RuntimeError("No SECONDARY STA found.")

# Suppress deprecation and other warnings for cleaner output
warnings.filterwarnings('ignore', category=DeprecationWarning)
warnings.filterwarnings('ignore')

path_stacks_folder = Path("STA")

sta_primary = get_primary(path_stacks_folder)
sta_secondary = get_first_secondary(path_stacks_folder, sta_primary)

print(f"[INFO] Using PRIMARY   : {sta_primary.name}")
print(f"[INFO] Using SECONDARY : {sta_secondary.name}")

product_primary = BiomassProduct.BiomassProductSTA(sta_primary)
product_secondary = BiomassProduct.BiomassProductSTA(sta_secondary)

# Access and print the noDataValue from the primary product's annotation
noDataValue=product_primary.noDataValue
print("NoData Value (primary product):", noDataValue)
# Check if waveNumbers LUT is loaded
if hasattr(product_primary, "coregistration_waveNumbers") and hasattr(product_secondary, "coregistration_waveNumbers"):

    waveNumbers_primary = product_primary.coregistration_waveNumbers
    waveNumbers_secondary = product_secondary.coregistration_waveNumbers

    # Compute stats
    waveNumbers_primary_avg = np.mean(waveNumbers_primary)
    waveNumbers_primary_std = np.std(waveNumbers_primary)
    waveNumbers_secondary_avg = np.mean(waveNumbers_secondary)
    waveNumbers_secondary_std = np.std(waveNumbers_secondary)

    # Plot settings
    std_multiplier = 3
    aspect_ratio ='auto' # 0.5
    colorbar_fraction = 0.05
    colorbar_pad = 0.04

    plt.close('all')
    fig, axes = plt.subplots(1,2) 
    ax1 = axes[0].imshow(
        waveNumbers_primary, 
        cmap='rainbow',
        aspect=aspect_ratio, 
        vmin =  waveNumbers_primary_avg - std_multiplier * waveNumbers_primary_std,
        vmax =  waveNumbers_primary_avg + std_multiplier * waveNumbers_primary_std
    )
    axes[0].set_title('waveNumbers primary')


    ax2 = axes[1].imshow(
        waveNumbers_secondary, 
        cmap='rainbow',
        aspect=aspect_ratio,
        vmin = waveNumbers_secondary_avg - std_multiplier *waveNumbers_secondary_std,
        vmax = waveNumbers_secondary_avg + std_multiplier *waveNumbers_secondary_std
    )
    axes[1].set_title('waveNumbers secondary')
    axes[1].yaxis.set_tick_params(labelleft=False)

    fig.colorbar(ax1, ax=axes[0], orientation='vertical', fraction=colorbar_fraction, pad=colorbar_pad)
    fig.colorbar(ax2, ax=axes[1], orientation='vertical', fraction=colorbar_fraction, pad=colorbar_pad)
    fig.tight_layout()
    plt.show()
else:
    print("❌ coregistration LUT or required variables not found in product.")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].hist(waveNumbers_primary.values.ravel(), bins='auto', alpha=0.8)
axes[0].set_title("waveNumbers_primary Histogram")
axes[0].set_xlabel("Value")
axes[0].set_ylabel("Frequency")
axes[0].grid(True)

axes[1].hist(waveNumbers_secondary.values.ravel(), bins='auto', alpha=0.8)
axes[1].set_title("waveNumbers_secondary Histogram")
axes[1].set_xlabel("Value")
axes[1].set_ylabel("Frequency")
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

plot FH and its quality layer

```python
# get the rasters
nan_value=product_fh_l2a.floatNoDataValue

with rasterio.open(l2a_fh) as fh_raster:
    fh = fh_raster.read(1)
    
    # Get the coordinates for the plot
    transform = fh_raster.transform
    width = fh_raster.width
    height = fh_raster.height
    cols, rows = np.meshgrid(np.arange(width), np.arange(height))
    xs, ys = rasterio.transform.xy(transform, rows, cols)
    
    # Convert to numpy arrays
    lons = np.array(xs)
    lats = np.array(ys)



with rasterio.open(l2a_fh_quality) as fh_quality_raster:
    fh_quality = fh_quality_raster.read(1)

fh_avg = np.mean(np.ma.masked_equal(fh,nan_value))
fh_std = np.std(np.ma.masked_equal(fh, nan_value))

fh_quality_avg = np.mean(np.ma.masked_equal(fh_quality, nan_value))
fh_quality_std = np.std(np.ma.masked_equal(fh_quality, nan_value))
```

```python
import os
std_multiplier = 3
aspect_ratio = 0.7
colorbar_fraction = 0.05
colorbar_pad = 0.04


fig, axes = plt.subplots(1,2) 
ax1 = axes[0].imshow(
    np.ma.masked_equal(fh,nan_value), 
    cmap='YlGn',
    aspect=aspect_ratio, 
    extent=(lons.min(), lons.max(), lats.min(), lats.max()),
    vmin =  fh_avg - std_multiplier * fh_std,
    vmax =  fh_avg + std_multiplier * fh_std, 
    interpolation= "nearest"
)
axes[0].set_title('FH [m]')
axes[0].set_xlabel("Lon (wgs84)")
axes[0].set_ylabel("Lat (wgs84)")

ax2 = axes[1].imshow(
    np.ma.masked_equal(fh_quality, nan_value), 
    cmap='rainbow_r',
    aspect=aspect_ratio,
    extent=(lons.min(), lons.max(), lats.min(), lats.max()),
    vmin = 0,
    vmax = 100
)
axes[1].set_title('FH quality []')
axes[1].yaxis.set_tick_params(labelleft=False)
axes[1].set_xlabel("Lon (wgs84)")

fig.colorbar(ax1, ax=axes[0], orientation='vertical', fraction=colorbar_fraction, pad=colorbar_pad)
fig.colorbar(ax2, ax=axes[1], orientation='vertical', fraction=colorbar_fraction, pad=colorbar_pad)
fig.tight_layout()



plt.show()
plt.close
```
